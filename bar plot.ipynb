```python
import numpy as np
import matplotlib.pyplot as plt
 
  
# creating the dataset
data = {'CNN classifier':95.8, 'CNN-SVM classifier':99.19, 'CNN-MLP classifier':100, 'CNN-RF classifier':100, 
        'CNN-KNN classifier':100, 'CNN-XGBoost classifier':99.76}
classifier = list(data.keys())
accuracy = list(data.values())
  
fig = plt.figure(figsize = (12, 8))
 
# creating the bar plot
plt.bar(classifier, accuracy, color ='cornflowerblue', width = 0.3)
plt.grid(color='#95a5a6', linestyle='--', linewidth=2, axis='y', alpha=0.7)
plt.xlabel("Models")
plt.ylabel("Accuracy (%)")
plt.annotate("95.8", ('CNN classifier',96.2),fontsize=12, ha='center')
plt.annotate("99.19", ('CNN-SVM classifier',99.5),fontsize=12, ha='center')
plt.annotate("100", ('CNN-MLP classifier',100.3),fontsize=12, ha='center')
plt.annotate("100", ('CNN-RF classifier',100.3),fontsize=12, ha='center')
plt.annotate("100", ('CNN-KNN classifier',100.3),fontsize=12, ha='center')
plt.annotate("99.76", ('CNN-XGBoost classifier',100.2),fontsize=12, ha='center')
plt.ylim([90, 101])
plt.show()
```
